```python
workplace_classes_age=pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Population/Workplace/WorkplaceClassAge.csv").rename(columns={'Territorio':'Province', 'Ateco 2007':'Ateco3_2007', 'Classe di età':'Age_Group', 'Value':'Population', "D3":"ATECO3_2007"})
workplace_classes_age=workplace_classes_age.drop(columns=["Seleziona periodo", "Ateco3_2007", "Tipo dato", "Flag Codes", "Flags", "D1", "D2", "D6", "D9"])
workplace_classes_age["Population"]=[int(value) for value in workplace_classes_age["Population"]]
workplace_classes_age=workplace_classes_age[workplace_classes_age["Province"] != "Emilia-Romagna"]
workplace_classes_age=workplace_classes_age[workplace_classes_age["Province"] != "Friuli-Venezia Giulia"]
workplace_classes_age=workplace_classes_age[workplace_classes_age["Province"] != "Lazio"]
workplace_classes_age=workplace_classes_age[workplace_classes_age["ATECO3_2007"] != "0010"]
workplace_classes_age["Code"]= [provincial_dict[prov] for prov in workplace_classes_age["Province"]]
workplace_classes_age=workplace_classes_age.sort_values(by=["Code", "Age_Group", "ATECO3_2007",])
workplace_classes_age=workplace_classes_age[["Code", "Province", "ATECO3_2007", "Age_Group", "Population"]]
workplace_classes_age.to_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Population/Workplace/WorkplaceClassAge2017.csv", index=False)
```

```python
provincial_age_data=np.array(pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Population/AgeStratifiedProvincialPopulation2020-2017.csv"))

provincial_age_list=[]
for i in range(107):
    for j in range(3):
        for k in range(17):
            provincial_age_list.append(provincial_age_data[j][i])
```

```python
workplace_classes_age["2020/2017"]=provincial_age_list
workplace_classes_age["Population_2020"]=[int(i) for i in workplace_classes_age["2020/2017"]*workplace_classes_age["Population_2017"]]
workplace_classes_age=workplace_classes_age.sort_values(by=["Code","ATECO3_2007", "Age_Group",])
```
