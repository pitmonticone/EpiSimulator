```python
# Load mobility data 
mobility_data_phase1 = pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Mobility/Flow1.csv").rename(columns={'origin':'source', 'destination':'target', 'flow':'weight'})
# Create mobility edge-list
mobility_edgelist_phase1 = [tuple(weighted_edge) for weighted_edge in mobility_data_phase1.to_numpy()]
# Build mobility graph (weighted directed)
mobility_graph_phase1 = nx.DiGraph()
mobility_graph_phase1.add_weighted_edges_from(mobility_edgelist_phase1)
# Save mobility graph data
mobility_data_phase1.to_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Mobility/Flow1EdgeList.csv",index=False)
nx.write_graphml(mobility_graph_phase1, "/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/GraphML/Flow1.graphml")
nx.write_edgelist(mobility_graph_phase1, "/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/EdgeLists/Flow1.edgelist")

# ADD MOBILITY DATA PATHS 
### PIETRO1: "/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Pepe2020/flow.csv"
### PIETRO2: "/Users/pietromonticone/github/DigitalEpidemiologyProject/Data/CSV/2020/Pepe2020/flow.csv"
```

### Population Data 

```python
# Load population data 
provincial_demographics = pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Population/ProvincialDemographics.csv").rename(columns = {'Codice provincia': 'code', 'Provincia': 'name', 'Età': 'age'})
# Add total population per province
provincial_demographics["population"] = provincial_demographics["Totale Maschi"] + provincial_demographics["Totale Femmine"]
# Remove useless fields
provincial_demographics = provincial_demographics.drop(['age','Maschi celibi', 'Maschi coniugati', 'Maschi divorziati', 'Maschi vedovi', 'Maschi uniti civilmente', 'Maschi già in unione civile (per scioglimento)', 'Maschi già in unione civile (per decesso del partner)', 'Totale Maschi','Femmine nubili', 'Femmine coniugate', 'Femmine divorziate', 'Femmine vedove', 'Femmine unite civilmente', 'Femmine già in unione civile (per scioglimento)', 'Femmine già in unione civile (per decesso del partner)', 'Totale Femmine'], axis=1)

# Load administrative data
provinces = pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Pepe2020/Provinces.csv").rename(columns = {'COD_PROV': 'code', 'DEN_PCM': 'name'}).drop(['INDEX', 'SIGLA'], axis=1)
# Add population data
provinces['population'] = list(provincial_demographics.groupby('code').sum()['population'])

# Save population data
###provinces.to_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/ProvincialPopulation.csv", index=False)
###provincial_demographics.to_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/AgeSpecificProvincialPopulation.csv", index=False)

# ADD POPULATION DATA PATHS 
### PIETRO1: "/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/ProvincialDemographics.csv"
### PIETRO2: "/Users/pietromonticone/gihtub/DigitalEpidemiologyProject/Data/CSV/2020/ProvincialDemographics.csv"
```

### Graph Metadata

#### Phase 1

```python
# Add name metadata
mobility_graph_phase1.name = 'Provincial Mobility Graph of Italy'
# Add population metadata
mobility_graph_phase1.population = {}
for node in mobility_graph_phase1.nodes:
    mobility_graph_phase1.population[node] = int(provinces[provinces['code']==node].population)
nx.set_node_attributes(mobility_graph_phase1, mobility_graph_phase1.population, 'population') 

# Initialize geo-spatial metadata 
mobility_graph_phase1.position = {}
# Load geo-spatial data
provincial_geospatial = gpd.read_file("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/Shapefiles/ProvCM01012020/ProvCM01012020_WGS84.shp")
# Add geo-spatial metadata
for node in mobility_graph_phase1.nodes:
    mobility_graph_phase1.position[node] = (float(provincial_geospatial[provincial_geospatial['COD_PROV']==node].centroid.x), float(provincial_geospatial[provincial_geospatial['COD_PROV']==node].centroid.y))
nx.set_node_attributes(mobility_graph_phase1, mobility_graph_phase1.position, 'position') 

# ADD GEOSPATIAL DATA PATHS 
### PIETRO1: "/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/Shapefiles/ProvCM01012020/ProvCM01012020_WGS84.shp"
### PIETRO2: "/Users/pietromonticone/github/DigitalEpidemiologyProject/Data/Shapefiles/ProvCM01012020/ProvCM01012020_WGS84.shp"
```

## Visualization

#### Phase 1

```python
# Showcase connectivity
#print("CONNECTIVITY:\n•",mobility_graph.number_of_nodes(),"nodes;\n•",mobility_graph.number_of_edges(),"edges.")

# Create nodal population list
population_list = [10000*mobility_graph_phase1.population[i] for i in list(mobility_graph_phase1.nodes)] / sum([mobility_graph_phase1.population[i] for i in list(mobility_graph_phase1.nodes)])

# Visualize Graph
fig=plt.figure(figsize=(35,35))
nx.draw_networkx_nodes(mobility_graph_phase1, 
                       pos=mobility_graph_phase1.position,
                       node_size=population_list) #list(mobility_graph.population)
nx.draw_networkx_edges(mobility_graph_phase1,
                       pos=mobility_graph_phase1.position, 
                       arrowstyle='->',
                       alpha = 0.6, 
                       width = 50*np.array(list(nx.get_edge_attributes(mobility_graph_phase1,'weight').values())))
plt.axis('off')
# Visualize the contact matrix 
plt.title("Phase 1", fontsize=80)
plt.savefig("/Users/Pit/GitHub/DigitalEpidemiologyProject/Images/Mobility/Phase1.png", dpi=200)
plt.show()
```
