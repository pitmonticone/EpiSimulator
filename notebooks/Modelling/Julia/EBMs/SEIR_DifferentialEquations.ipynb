# SEIR Model using `DifferentialEquations`

The implementation in this section considers proportions of susceptibles, exposed, infectious individuals in an open population, with no additional mortality associated with infection (such that the population size remains constant and R is not modelled explicitly).

$$ \dot{S} = \mu - \beta S I - \mu S $$
$$ \dot{E} = \beta S E - (\sigma + \mu) E $$
$$ \dot{I} = \sigma E - (\gamma + \mu) I $$
$$ \dot{R} = \gamma I = \mu R $$

## Packages

```julia
# Numerical Computation
using DifferentialEquations

# Data Visualization
using Plots
using PlotThemes
using PyPlot
```

## Flow

```julia
function Œ¶!(du,u,p,t)
    S,E,I,R = u
    Œ≤,œÉ,Œ≥,Œº = p
    
    du[1] = Œº-Œ≤*S*I-Œº*S
    du[2] = Œ≤*S*I-(œÉ+Œº)*E
    du[3] = œÉ*E - (Œ≥+Œº)*I
    du[4] = Œ≥*I
end
```

## Parameters

```julia
ùí´ = [520/365,1/60,1/30,774835/(65640000*365)]      # Model Parameters
‚Ñ¨ = [0.8,0.1,0.1, 0.0]                             # Initial condition
ùíØ = (0.0,365.0)                                    # Time 
```

## Solution

```julia
# Problem Definition
problem = ODEProblem(Œ¶!, ‚Ñ¨, ùíØ, ùí´)

# Problem Solution
solution = solve(problem);
```

## Visualization 

```julia
# Select Theme
theme(:default)

plot(solution,
     title  = "SEIR Evolution", 
     xlabel = "Time", 
     ylabel = "Fraction", 
     label=["S" "E" "I" "R"],
     lw=2,
     legend = :right)
```

```julia
# Panel Plot
plot(solution,
     title  = ["Susceptible" "Exposed" "Infected" "Recovered"], 
     xlabel = "Time", 
     ylabel = "Fraction", 
     label=false,
     layout = (2,2),
     lw=2,
     legend = :right)
```

```julia
# Phase Diagram
plot(solution, plotdensity=1000, 
    vars=(1,2,3),
    lw=2.5,
    label = false,
    #xlabel = "S",
    xlabel = "S", 
    ylabel = "E",
    zlabel = "I")
```
