# SIR Model using `Agents`

The agent-based model approach implemented here is:

* Stochastic
* Discrete in time
* Discrete in state

### Packages

```julia
# Data Management 
using DataFrames
using DrWatson

# Statistics
using Random
using Distributions

# Modelling
using Agents

# Data Visualization
using Plots
using PlotThemes
```

### Utility Functions

```julia
function rate_to_proportion(r::Float64,t::Float64)
    1-exp(-r*t)
end;

# Surveillance 
susceptible(x) = count(i == :S for i in x)
infected(x) = count(i == :I for i in x)
recovered(x) = count(i == :R for i in x);
```

### Agent Type 

```julia
# Agent Definition
mutable struct Person <: AbstractAgent
    id::Int64
    status::Symbol 
end
```

### Dynamics

#### Micro / Agent-Level

```julia
# Transmission Dynamics
function transmit!(agent, model)
    # If I'm not susceptible, I return
    agent.status != :S && return
    ncontacts = rand(Poisson(model.properties[:c]))
    for i in 1:ncontacts
        # Choose random individual
        alter = random_agent(model)
        if alter.status == :I && (rand() ≤ model.properties[:β])
            # An infection occurs
            agent.status = :I
            break
        end
    end
end;

# Recovery Dynamics
function recover!(agent, model)
    agent.status != :I && return
    if rand() ≤ model.properties[:γ]
            agent.status = :R
    end
end;

function agent_step!(agent, model)
    transmit!(agent, model)
    recover!(agent, model)
end;
```

#### Macro / System-Level

```julia
# Model Initialization
function init_model(β::Float64,c::Float64,γ::Float64,N::Int64,I0::Int64)
    properties = @dict(β,c,γ)
    model = ABM(Person; properties=properties)
    for i in 1:N
        if i <= I0
            s = :I
        else
            s = :S
        end
        p = Person(i,s)
        p = add_agent!(p,model)
    end
    return model
end;
```

### Parameters

```julia
# Time 
δt = 0.1
nsteps = 400
tf = nsteps*δt
t = 0:δt:tf;

# Model Parameters
β = 0.05                           # Transmission rate 
c = 10.0*δt                        # Contact rate 
γ = rate_to_proportion(0.25,δt);   # Mortality rate

# Initial Conditions
N = 1000
I0 = 10;
```

### Solution

```julia
# Seed Selection
Random.seed!(1234);

# Model Instantiation
abm_model = init_model(β,c,γ,N,I0)

# Data Collection
to_collect = [(:status, f) for f in (susceptible, infected, recovered)]
data, _ = run!(abm_model, agent_step!, nsteps; adata = to_collect);
```

### Visualization

```julia
# Data Manipulation
data[!,:t] = t;

# Select Theme
theme(:default)

# Plot
plot(t,data[:,2],label="S",
     xlab="Time",
     ylabel="Number",
     title="SIR ABM", 
     legend=:topright)
plot!(t,data[:,3],label="I")
plot!(t, data[:,4],label="R")
```
