## Regional Level

### Data Manipulation 

```python
# Store GeoDataFrame
regional_data = gpd.read_file(regions)

# Fill column of neighbors 
depi.expand(data=regional_data, level="regional")
    
# Visualize GeoDataFrame
regional_data.plot(figsize=(10, 10)).set_axis_off()
```

### Graph Data Manipulation  

```python
# Build the regional undirected graph
regional_graph = depi.build_graph(data = regional_data,
                                  level = "regional",
                                  graph = nx.Graph())
```

### Graph Data Visualization 

```python
# Visualize graph 
fig=plt.figure(figsize=(15,8))
nx.draw_networkx_nodes(regional_graph, pos=regional_graph.pos,node_size=80)
nx.draw_networkx_edges(regional_graph, pos=regional_graph.pos) 
nx.draw_networkx_labels(regional_graph, pos=regional_graph.pos)
plt.axis('off')
plt.show()
```

## Provincial Level

## Municipal Level

### Data Manipulation 

```python
# Store GeoDataFrames
municipal_data = gpd.read_file(municipalities)
demographic_data = pd.read_csv(demographics)

# Fill column of neighbors 
depi.expand(data=municipal_data, level="municipal", demo = demographic_data)
    
# Visualize GeoDataFrame
municipal_data.plot(figsize=(10, 10)).set_axis_off()
```

### Graph Data Manipulation  

```python
# Build the regional undirected graph
municipal_graph = depi.build_graph(data = municipal_data,
                                  level = "municipal",
                                  graph = nx.Graph())
```

```python
nx.get_node_attributes(municipal_graph, 'population')
```

### Graph Data Visualization 

```python
# Showcase connectivity
print("CONNECTIVITY:\n•",municipal_graph.number_of_nodes(),"nodes;\n•",municipal_graph.number_of_edges(),"edges.")

# Visualize graph 
fig=plt.figure(figsize=(15,8))
nx.draw_networkx_nodes(municipal_graph, pos=municipal_graph.pos,node_size=0.4)
nx.draw_networkx_edges(municipal_graph, pos=municipal_graph.pos, alpha = 0.19) 
#nx.draw_networkx_labels(K, pos=K.pos)
plt.axis('off')
plt.show()
```
