## Household Size

```python
household_sizes=pd.read_csv("/Users/Pit/GitHub/DigitalEpidemiologyProject/Data/CSV/2020/Population/HouseholdSizes.csv")
household_sizes_2019=household_sizes[household_sizes.TIME ==2019]
household_sizes_2019=household_sizes_2019.drop(columns=["TIME","Flag Codes", "Flags", "Select time", "NUMEROSITA", "MISURA_AVQ"])
household_sizes_2019=household_sizes_2019[household_sizes_2019["Measure"] =="thousands value"]
household_sizes_2019["Value"]=[int(size) for size in household_sizes_2019["Value"]]
household_sizes_2019=household_sizes_2019.sort_values(by=["Number of household members"])
```
